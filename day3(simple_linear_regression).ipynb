```python
#import all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the dataset
dataset= pd.read_csv('studentscores.csv')
X = dataset.iloc[:, :1].values
Y = dataset.iloc[:,1].values
```

```python
import sklearn
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=1/4, random_state=0)
```

```python
from sklearn.model_selection import train_test_split  #cross_validation has been replaced with model_selection
# cross_validation can no longer be used
X_train, X_test, Y_train ,Y_test = train_test_split(X,Y, test_size =1/4, random_state = 0)
```

```python
#make a regressor to fit data into x_train and y_train
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

```python
#predicting the result
Y_pred = regressor.predict(X_test)
```

```python
#visualizing the trainig result
plt.scatter(X_train , Y_train , color= 'red')
plt.plot(X_train , regressor.predict(X_train),color='blue')
```

```python
#visualizing the training result
plt.scatter(X_test, Y_test , color= 'red')
plt.plot(X_test, regressor.predict(X_test), color='blue')
```
